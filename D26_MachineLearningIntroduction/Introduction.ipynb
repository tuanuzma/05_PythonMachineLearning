# Machine Learning Work Flow

```python
# pip install scikit-learn
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from matplotlib import pyplot as plt
from IPython.display import display, HTML
```

### Data Acquisition

```python
# Let us create the data
X = np.arange(10).reshape((5, 2))
X
```

```python
y = range(5)
y
```

```python
myDataFrame = pd.DataFrame(X, columns=['X1', 'X2'])
myDataFrame['y'] = y
myDataFrame
```

```python
plt.figure(figsize=(4, 3))
scatter = plt.scatter(X[:, 0], X[:, 1], c=y)
plt.colorbar(scatter)
```

### Data Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
def side_by_side(*dfs):
    html = '<div style="display:flex">'
    for df in dfs:
        html += '<div style="margin-right: 2em">'
        html += df.to_html()
        html += '</div>'
    html += '</div>'
    display(HTML(html))
```

```python
train = pd.DataFrame(X_train, columns=['X1', 'X2'])
train['y'] = y_train
test = pd.DataFrame(X_test, columns=['X1', 'X2'])
test['y'] = y_test
side_by_side(train, test)
```

### Train the Model

```python
# Let us create an object of the Machine Learning Algorithm
model = LinearRegression()
model
```

```python
# to train an algorithm the method used is fit
model.fit(X_train, y_train)
```

### Test the Model

```python
prediction = model.predict(X_test)
prediction
```
